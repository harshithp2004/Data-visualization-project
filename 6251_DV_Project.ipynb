Univariate Analysis

1.What does the line plot of 'Happiness Score' for the top 50 countries by rank reveal about the variation in happiness scores among these countries, and are there any noticeable trends or patterns?

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with the 'Happiness score', 'Country', and 'Rank' columns
top_50_df = df[df['RANK'] <= 50]  # Select only the top 50 countries by rank
countries = top_50_df['Country']
happiness_scores = top_50_df['Happiness score']

plt.figure(figsize=(12, 6))
plt.plot(countries, happiness_scores, marker='o', linestyle='-')
plt.title('Line Plot of Happiness Score for the Top 50 Countries by Rank')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()

```

Conclusion:The line plot reveals variations in 'Happiness Score' among the top 50 ranked countries. Some exhibit stable scores, while others show fluctuations, indicating diverse happiness trends.

5.Is there any relationship between Happiness score and Corruption

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sort the DataFrame by 'Happiness score' in descending order
df_sorted = df.sort_values(by='Happiness score', ascending=False)

# Select the top 20 and bottom 20 countries
top_20_countries = df_sorted.head(20)
bottom_20_countries = df_sorted.tail(20)

# Create subplots with two rows and one column
fig, axes = plt.subplots(2, 1, figsize=(10, 12))
sns.set(style="whitegrid")

# Top 20 countries subplot
sns.scatterplot(ax=axes[0], x=top_20_countries['Happiness score'], y=top_20_countries['Explained by: Perceptions of corruption'], hue=top_20_countries['Country'], palette='Set2')
sns.kdeplot(ax=axes[0], x=top_20_countries['Happiness score'], y=top_20_countries['Explained by: Perceptions of corruption'], cmap='coolwarm', fill=True, levels=20)
axes[0].set_title('Top 20 Countries: Happiness Score vs. Perceptions of Corruption')
axes[0].set_xlabel('Happiness Score')
axes[0].set_ylabel('Perceptions of Corruption')
axes[0].legend(title='Country', loc='upper right')

# Bottom 20 countries subplot
sns.scatterplot(ax=axes[1], x=bottom_20_countries['Happiness score'], y=bottom_20_countries['Explained by: Perceptions of corruption'], hue=bottom_20_countries['Country'], palette='Set2')
sns.kdeplot(ax=axes[1], x=bottom_20_countries['Happiness score'], y=bottom_20_countries['Explained by: Perceptions of corruption'], cmap='coolwarm', fill=True, levels=20)
axes[1].set_title('Bottom 20 Countries: Happiness Score vs. Perceptions of Corruption')
axes[1].set_xlabel('Happiness Score')
axes[1].set_ylabel('Perceptions of Corruption')
axes[1].legend(title='Country', loc='upper right')

# Adjust space between subplots
plt.tight_layout()
plt.show()

```

Conclusion:The dual subplot analysis contrasts the relationship between 'Happiness Score' and 'Perceptions of Corruption' for the top 20 and bottom 20 countries. The scatterplots indicate varying associations, while the KDE plots show the distribution of data points, revealing potential patterns in how corruption perceptions affect happiness across different countries.

4.How does the visual comparison of factor contributions (excluding Generosity) for the top 20 and bottom 20 ranked countries, as presented in the side-by-side stacked area plots, offer insights into the disparities and trends among these distinct groups of nations?

```python
import matplotlib.pyplot as plt
import pandas as pd

sorted_df = df.sort_values(by="RANK")

# Select the top 20 and bottom 20 countries and relevant columns (excluding 'Generosity')
top_20_countries = sorted_df.head(20)
bottom_20_countries = sorted_df.tail(20)
selected_columns = ['Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy']

top_20_data = top_20_countries[selected_columns].T
bottom_20_data = bottom_20_countries[selected_columns].T

x_labels_top_20 = top_20_countries["Country"]
x_labels_bottom_20 = bottom_20_countries["Country"]

# Create subplots with two stack plots side by side
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Stack plot for the top 20 countries
axes[0].stackplot(x_labels_top_20, top_20_data, labels=top_20_countries['Country'])
axes[0].set_xlabel("Country")
axes[0].set_ylabel('Contribution')
axes[0].set_title('Stack Plot of Factors for Top 20 Countries (Excluding Generosity)')
axes[0].legend(loc="upper left", bbox_to_anchor=(1, 1))
axes[0].tick_params(axis='x', rotation=45)

# Stack plot for the bottom 20 countries
axes[1].stackplot(x_labels_bottom_20, bottom_20_data, labels=bottom_20_countries["Country"])
axes[1].set_xlabel("Country")
axes[1].set_ylabel('Contribution')
axes[1].set_title('Stack Plot of Factors for Bottom 20 Countries (Excluding Generosity)')
axes[1].legend(loc="upper left", bbox_to_anchor=(1, 1))
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

```

Conclusion: The stacked area plots, showcasing factor contributions (excluding Generosity) for the top 20 and bottom 20 ranked countries. This visual comparison aids in discerning variations in influential factors between these two groups, facilitating insightful analysis of the dataset.
